```python
dt = pd.read_csv("../outputs/scores/0_dataset.csv")
explanator = EBMExplanator("../Models/GlassBox/ebm/saved_models/EBMReg")
```

```python
def plot_single_function(fun:pd.DataFrame, name:str):
    f, axs = plt.subplots(figsize=(5,3))
    sns.lineplot(x="x",y="y",data=fun, label=name, ax=axs)
    axs.fill_between(fun['x'], fun['lower'], fun['upper'], color='#4B4BFF', alpha=0.3)
    axs.grid(color='grey', linestyle='-.', linewidth=0.5, alpha=0.5)
    axs.set_xlabel("fitness value", fontsize=10)
    axs.set_ylabel("Contribute", fontsize=12)
    axs.legend(fontsize=8)
    plt.show()
```

```python
fun = explanator.piecewise_functions["fitness_comp_sim_basic"].show_function()
```

```python
plot_single_function(fun,"fitness_comp_sim_basic")
```
