```python
import pandas as pd
from InterpretabilityByDesign.EBMExplanator import EBMExplanator
import matplotlib.pyplot as  plt
import seaborn as sns

sns.set_style("darkgrid")
import warnings
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")
%reload_ext autoreload
%autoreload 2
```

```python
dt = pd.read_csv("../outputs/scores/0_dataset.csv")
explanator = EBMExplanator("../Models/GlassBox/ebm/saved_models/EBMReg",
                           dataset_path=f"./../outputs/scores/0_dataset_tr.csv")
```

## fitness education

```python
edu_basic = explanator.piecewise_functions["fitness_edu_basic"].show_function()
edu_basic_points = explanator.piecewise_functions["fitness_edu_basic"].points
edu_bonus = explanator.piecewise_functions["fitness_edu_bonus"].show_function()
edu_bonus_points = explanator.piecewise_functions["fitness_edu_bonus"].points
```

```python
_, axs = plt.subplots(ncols=2,figsize=(12,4),sharey=True,sharex=True)
sns.lineplot(x="x",y="y",data=edu_basic, ax=axs[0])
sns.scatterplot(x="x",y="y",data=edu_basic_points, ax=axs[0], color="#8B0000", label="Unique values")
axs[0].fill_between(edu_basic['x'], edu_basic['lower'], edu_basic['upper'], color='#4B4BFF', alpha=0.3)
axs[0].grid(color='grey', linestyle='-.', linewidth=0.5, alpha=0.5)
axs[0].set_xlabel("fitness value", fontsize=10)
axs[0].set_ylabel("Contribute", fontsize=12)
axs[0].set_title("fitness edu_basic")
axs[0].text(0.2, edu_basic_points.values[1][1] + 0.1 , "not satisfy",
            ha='center', va='bottom', fontsize=16)
axs[0].text(0.8, edu_basic_points.values[0][1] + 0.1 , "satisfy",
            ha='center', va='bottom', fontsize=16)
axs[0].set_ylim([-8,10])


sns.lineplot(x="x",y="y",data=edu_bonus, ax=axs[1])
sns.scatterplot(x="x",y="y",data=edu_bonus_points, ax=axs[1], color="#8B0000", label="Unique values")
axs[1].fill_between(edu_bonus['x'], edu_bonus['lower'], edu_bonus['upper'], color='#4B4BFF', alpha=0.3)
axs[1].grid(color='grey', linestyle='-.', linewidth=0.5, alpha=0.5)
axs[1].set_xlabel("fitness value", fontsize=10)
axs[1].set_ylabel("Contribute", fontsize=12)
axs[1].set_title("fitness edu_bonus")
axs[1].text(0.2, edu_bonus_points.values[0][1] + 0.1, "not satisfy",
            ha='center', va='bottom', fontsize=16)
axs[1].text(0.8, edu_bonus_points.values[1][1] + 0.1 , "satisfy",
            ha='center', va='bottom', fontsize=16)
axs[1].set_ylim([-8,10])

plt.suptitle("Fitness Education",fontsize=18)
plt.show()
```

## fitness city and age

```python
city = explanator.piecewise_functions["fitness_city"].show_function()
city_points = explanator.piecewise_functions["fitness_city"].points
age = explanator.piecewise_functions["fitness_age"].show_function()
age_points = explanator.piecewise_functions["fitness_age"].points
```
